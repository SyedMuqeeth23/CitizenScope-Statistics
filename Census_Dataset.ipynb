**Q.4) Calculate state-wise : A. Total number of population.**

```python
df.head(1)
```

```python
df.groupby('State_name').Population.sum()
```

```python
df.groupby('State_name').Population.sum().sort_values()
```

```python
df.groupby('State_name').Population.sum().sort_values(ascending = False)
```

**B. Total no. of the population with different religions.**

```python
df.columns
```

```python
df.groupby('State_name')['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains'].sum().sort_values(by = 'Muslims')
```
